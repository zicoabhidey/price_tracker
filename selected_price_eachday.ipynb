### Start running from the next cell if you are running for the first time

```python
def ssd(myurls):
    for url in myurls:
        headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
        r = requests.get(url, headers=headers)#, proxies=proxies
        content = r.content
        soup = BeautifulSoup(content)
        product_name = soup.find('span', attrs={'id':'productTitle'}).text.strip()[:-1]
        productName.append(product_name)
        price = soup.find('span', attrs={'id':'priceblock_ourprice'}).text[1:6]
        priceList.append(price)
        rating = soup.find('span', attrs={'class':'a-icon-alt'}).text[:3]
        ratings.append(rating)
        numRating = soup.find('span', attrs={'id':'acrCustomerReviewText'}).text.strip(' ratings')
        numRatings.append(numRating)
        availability = soup.find('div', attrs={'id':'availability'}).text.strip().replace('.','').rstrip('\n        \n        \n    \nOrder it now')
        availabilities.append(availability)
    return productName, priceList, ratings, numRatings, availabilities
```
